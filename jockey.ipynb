```python
# Configure the client with the correct base URL and API key
client = get_client()

# List all assistants
assistants = await client.assistants.search()
```

```python
# We auto-create an assistant for each graph you register in config.
jockey: Assistant = assistants[0]
```

```python
# Start a new thread
thread = await client.threads.create()
```

```python
index_input = "Using index 66f1b1c8163dbc55ba3bb1b6."
story_board_input = "Please search for 3 clips, return an image for each clip. No need to generate the video."
```

```python
user_input = "I am trying to create a compelling 1-minute video describing why the degree of difficulty in an Olympic Gymnastics event can impact the judge's score as much as the execution of the routine."
```

```python
MAX_RETRIES = 3
RESET_DELAY = 2

retry_count = 0
success = False
output_events = []
```

```python

while not success and retry_count < MAX_RETRIES:
    try:
        # Start a streaming run
        chat_history = [HumanMessage(content=index_input + user_input + story_board_input, name="user")]
        jockey_input = {
            "chat_history": chat_history,
            "made_plan": False,
            "next_worker": None,
            "active_plan": None,
        }

        # Start the streaming run
        async for event in client.runs.stream(
            thread["thread_id"], 
            jockey["assistant_id"], 
            input=jockey_input, 
            stream_mode="values"
        ):
            print(event)
            output_events.append(event)

        success = True

    except Exception as e:
        print(f"Error: {e}")
        retry_count += 1
        if retry_count < MAX_RETRIES:
            print(f"Retrying... ({retry_count}/{MAX_RETRIES})")
            await asyncio.sleep(RESET_DELAY)
        else:
            print("Max retries exceeded. Exiting...")
```

```python
chat_history = [HumanMessage(content=index_input + user_input + story_board_input, name="user")]
jockey_input = {
    "chat_history": chat_history,
    "made_plan": False,
    "next_worker": None,
    "active_plan": None,
}

# Start the streaming run
async for event in client.runs.stream(
    thread["thread_id"], 
    jockey["assistant_id"], 
    input=jockey_input, 
    stream_mode="values"
):
    print(event)
    output_events.append(event)
```

```python
def extract_url(content):
    pattern = r'\((https:\/\/[\s\S]*?)\)'
    print(f"Extracting...")

    # Find all matches in the msg content
    raw_urls = re.findall(pattern, content)

    # Clean up
    urls = []
    for url in raw_urls:
        clean_url = ''.join(url.split())
        urls.append(clean_url)

    # for url in raw_urls:
    #     print(f"Found {url}")
    return urls

def download_video(urls, start=1):

    if not urls:
        print("No thumbnail URLs found")
        return

    # Create a directory to save the images
    # current_dir = os.getcwd()
    # thumbnails_dir = os.path.join(current_dir, 'thumbnails')
    os.makedirs('thumbnails', exist_ok=True)

    # Download each image
    for idx, url in enumerate(urls, start=1):
        try:
            response = requests.get(url)
            response.raise_for_status()  # Check if the request was successful

            # Save the image with a filename like 'thumbnail1.jpeg'
            image_filename = f'thumbnails/thumbnail{idx}.jpeg'
            with open(image_filename, 'wb') as f:
                f.write(response.content)

            print(f"Downloaded {image_filename}")
        except requests.exceptions.RequestException as e:
            print(f"Failed to download image from {url}: {e}")
```

```python
urls = extract_url(assistant_message)
download_video(urls)
```
